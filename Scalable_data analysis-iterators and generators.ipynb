```python
import csv
import pandas as pd
import sys
import numpy as np
from datetime import date
import collections
import seaborn as sns
```

```python

def citibike2hod(filename):
    with open(filename,'r') as fi:
        reader = csv.DictReader(fi)
        for row in reader:
            if row['usertype']=='Subscriber':
                b_year= row['birth_year']
                if (b_year==' '):
                    del row
                else:
                    b_year_1=float(row['birth_year'])
                    age=date.today().year-b_year_1
                yield int(age)
                
count={}
for hod in citibike2hod("C:/Users/Darshika Rajasekaran/Downloads/citibike.csv"):
    count[hod] = count.get(hod,0)+1

    
```

```python
print(count)
data = {"age":count.keys(),"count":count.values()}
df = pd.DataFrame.from_dict(data,orient="index")
df=df.transpose()
df.describe()

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,15))
ax = sns.barplot(x = "age",y="count",data = df, ci = None)
```
